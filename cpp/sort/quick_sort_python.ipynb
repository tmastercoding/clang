```python
# function that partitions an array
# pivot = last element
# left side = smaller than pivot
# right side = greater than pivot

cnt = 0

def partition (a, start, end, side): 
    global cnt
    print( f'{side} side partition: {cnt} ') 
    print( a[start:end+1] )
    cnt += 1
    i = (start - 1)  
    pivot = a[end] # pivot element  
      
    for j in range(start, end):  
        # If current element is smaller than or equal to the pivot  
        if (a[j] <= pivot):  
            i = i + 1  
            print( f'i : {i}, j : {j}' )
            a[i], a[j] = a[j], a[i]  
      
    a[i+1], a[end] = a[end], a[i+1]  
  
    return (i + 1)  
      
# function to implement quick sort   
def quick(a, start, end, side): # a[] = array to be sorted, start = Starting index, end = Ending index      
    if (start < end):  
        p = partition(a, start, end, side ) # p is partitioning index  

        print("new pivot: ", p)
        # left side : letter than pivot
        quick(a, start, p - 1, 'left')  
        # right side : more than pivot
        quick(a, p + 1, end, 'right')  
  
          
def printArr(a): # function to print the array  
    for i in range(len(a)):  
        print (a[i], end = " ")
    print( )  
  
      
a = [ 70, 60, 50, 40, 30, 20 ,10 ]  
print("Before sorting array elements are - ")  
printArr(a)  
quick(a, 0, len(a)-1, 'init' )  
print("\nAfter sorting array elements are - ")  
printArr(a)  
```
