```python
# graph (X+1) by (Y+1)
# starts at 0, 0

# L red lines that end/ start on dots
# N stars on dots
# M yellow lines that connect 2 stars

# move stars to other dots
# to make yellow lines
# and red lines same
# (no overlap stars)
# (yellow lines retain connection)

# The distance stars moved = R
# Minimise R
# if star origin = (x, y)
# moved = (Δx, Δy)
# arrived in (x+ Δx), (y + Δy)
# R = sqrt((Δx)(Δx) + (Δy)(Δy))

### INPUT 
# X Y
# L
# L rows of (startX, startY) (endX, endY)
# N
# N rows of (x, y)
# M
# M rows of (startStar, endStar)

### OUTPUT
# N rows of where stars moved
```

```python
class redLine():
    def __init__(self, startX, startY, endX, endY):
        self.startX = startX
        self.startY = startY
        self.endX = endX
        self.endY = endY
        self.validDots = []
    
    def printSelf(self):
        print(self.startX, self.startY, self.endX, self.endY)
    
    def getValidDots(self):
        self.validDots = []
        if self.startX == self.endX:
            if self.startY > self.endY:
                self.validDots = list(map(lambda y: (self.startX, y), range(self.endY, self.startY+1)))
            else:
                self.validDots = list(map(lambda y: (self.startX, y), range(self.startY, self.endY+1)))
            return self.validDots
        elif self.startY == self.endY:
            if self.startX > self.endX:
                self.validDots = list(map(lambda x: (x, self.startY), range(self.endX, self.startX+1)))
            else:
                self.validDots = list(map(lambda x: (x, self.startY), range(self.startX, self.endX+1)))
            return self.validDots
        
        
        difX = abs(self.startX - self.endX)

        if self.startX > self.endX:
            posX = range(self.endX, self.startX+1)
            difY = self.startY - self.endY
        else:
            posX = range(self.startX, self.endX+1)
            difY = self.endY - self.startY

        slope = difY/difX
        intercept = self.startY - slope*self.startX

        for x in posX:
            y = x*slope + intercept
            if y.is_integer():
                self.validDots.append((x, int(y)))
        return self.validDots

redLines = []
# corners = set()
posDots = set()
for line in redInput:
    lineObj = redLine(*map(lambda x : int(x), line.split()))
    # lineObj.printSelf()
    # corners.add((lineObj.startX, lineObj.startY))
    # corners.add((lineObj.endX, lineObj.endY))
    posDots.update(lineObj.getValidDots())
    redLines.append(lineObj)
```

```python
from math import sqrt
class posDot():
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.connected = []
        self.connectedLen = 0
    
    def getConnect(self, redList):
        for red in redList:
            redConnect = [(red.startX, red.startY), (red.endX, red.endY)]
            if ((self.x, self.y)) in redConnect:
                self.connected.append(red)

        self.connectedLen = len(self.connected)
        return self.connected
    
    def getDist(self, dot2):
        difX = abs(self.x - int(dot2.x))
        difY = abs(self.y - int(dot2.y))

        return sqrt(difX*difX + difY*difY)


```
