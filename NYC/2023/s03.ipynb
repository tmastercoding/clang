```python
buffer = input()
inputSize = int(buffer.split()[0])
width = int(buffer.split()[1])

commands = {
    "<CENTER>": 0,
    "</CENTER>": -1,
    "<RIGHT>": 2,
    "</RIGHT>": -1
}

state = -1

for _ in range(inputSize):
    line = input()
    if line in commands.keys():
        state = commands[line]
        continue

    lines = []
    line = line.replace(' ', '-')
    lines.append(line)

    if len(line) > width:
        lines = line.split('-')
        size = len(lines)
        for idx in range(size):
            if (idx + 1) < size:
                temp = lines[idx] + '-' + lines[idx+1]
                if len(temp) <= width:
                    lines.pop(idx)
                    lines.pop(idx)
                    lines.insert(idx, temp)
                    break
                    
    
    if(state == -1):
        for line in lines:
            line += (width-len(line))*'-'
            print(line)
    
    if(state == 0):
        for line in lines:
            half = int((width - len(line))/2)

            line = (half)*'-' + line
            line += (width-len(line))*'-'
            print(line)
    
    if(state == 2):
        for line in lines:
            line = (width-len(line))*'-' + line
            print(line)
```

```python
text = \
'''Lorem ipsum
dolor sit amet
consectetur
adipiscing elit
Duis tristique
Nunc congue nisi
vitae suscipit tellus mauris
<CENTER>
Nunc congue nisi
vitae suscipit tellus mauris
Integer quis
Id venenati
a condimentum vitae
</CENTER>
<RIGHT>
Nisi est sit
amet facilisis magna
Volutpat ac tincidunt
vitae semper quis
</RIGHT>'''

width = 11

# width = input()
# width = int( width )
# text = []
# for _ in range( width ):
#     line = input()
#     text.append( line )

lines = text.split('\n')

def arrange( align: int, output: str, width: int ):

    length = len(output)

    if align == 0:
        rtn = output \
            + '-' * (width - length )
    elif align == 1:
        rtn = '-' * int((width - length )/2) \
            + output \
            + '-' * int((width - length )/2)
    elif align == 2:
        rtn =  '-' * (width - length ) \
        + output

    return rtn

align = 0

for line in lines:
    if line == '<CENTER>':
        align = 1
        continue
    elif line == '</CENTER>':
        align = 0
        continue
    elif line == '<RIGHT>':
        align = 2
        continue
    elif line == '</RIGHT>':
        align = 0
        continue

    if len( line ) <= width:
        line = line.replace(' ', '-')
        print( arrange( align, line, width ) )
    else:

        for word in line.split():
            print( arrange( align, word, width ) )

            
```
