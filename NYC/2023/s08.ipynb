```python
### Problem
# crossword n x m
# words are read:
# left to right
# up to down

# can make same word more than once
# cannot make 2 words 
# if one covers the other
# AND SAME DIRECTION

# words length sum = luck score
# maximsie luck score

### Input
# (N, M) crossword size
# Following N rows = crossword
# Number of lucky words
# Rows of lucky words
# 2 < word length <= min(N, M)

### OUTPUT
# Number of words made
# start + end pos of words
# e.g 1 3 6 3
```

```python
def inputGen(length, width, crossword, words):
    inputCase = str(length) + ' ' + str(width) + '\n'
    crossword = crossword.replace('\n', '')

    for idx in range(length):
        inputCase += crossword[idx*width: idx*width + width] + '\n'
    words = words.split('\n')
    inputCase += str(len(words)) + '\n'
    
    for word in words[:-1]:
        inputCase += word.strip() + '\n'
    inputCase+=words[-1]

    return inputCase

crossword = '''z
z
z
m
z
e
n
j
o
y
z
z
z
n
z
z
z
z
e
z
z
z
z
y
z'''
words = '''money
enjoy'''
case2 = inputGen(5, 5, crossword, words)
print(case2)
```

```python
cases = [case1]
for idx in range(1, 11):
    f = open(f"s08_cases/s08_case{idx}.txt")
    cases.append(f.read())
    f.close()
```

```python
from itertools import combinations
test = cases[2]

# split input to var
lines = test.split('\n')
length, width = map(lambda x: int(x), lines[0].split())
crossword = lines[1: 1+length]
wordSize = int(lines[1+length])
words = lines[2+length: ]

# class
class luckWord():
    def __init__(self, startPos, endPos, word):
        self.startPos = tuple(startPos)
        self.endPos = tuple(endPos)
        self.length = len(word)
        self.word = word
    
    def printSelf(self):
        print("start pos: " + str(self.startPos[0]) + ", "+ str(self.startPos[1]),\
             "end pos: " + str(self.endPos[0]) + ", "+ str(self.endPos[1]),\
             "length: " + str(self.length), self.word, sep = '\n')

foundWords = []

# get all words in crossword
for idx in range(length):
    for jdx in range(width):
        # longest words
        maxDownWord = luckWord((-1, -1), (-1, -1), '')
        maxRightWord = luckWord((-1, -1), (-1, -1), '')
        for word in words:
            wordLen = len(word)

            # up to down
            if (idx + wordLen) <= length and wordLen > maxDownWord.length:
                # potential 'word'
                # from point to point + wordLen
                temp = list(map(lambda x: crossword[idx+x][jdx], range(wordLen)))
                temp = ''.join(temp)

                # if same
                if temp == word:
                    maxDownWord = luckWord((idx, jdx), (idx+wordLen-1, jdx), word)

            # left to right
            if (jdx + wordLen) <= width and wordLen > maxRightWord.length:
                temp = list(map(lambda x: crossword[idx][jdx+x], range(wordLen)))
                temp = ''.join(temp)
                if temp == word:
                    maxRightWord = luckWord((idx, jdx), (idx, jdx+wordLen-1), word)
        
        if maxDownWord.length > 0:
            foundWords.append(maxDownWord)
        if maxRightWord.length > 0:
            foundWords.append(maxRightWord)

removeSet = set()

# get rid of overlap
for word, comp in combinations(foundWords, 2):
    if word == comp:
        continue
    
    # both directions
    for axis in range(2):
        if (word.startPos[axis] == word.endPos[axis])\
            and (comp.startPos[axis] == comp.endPos[axis])\
            and (word.startPos[axis] == comp.endPos[axis]):

            wordPos = set(range(word.startPos[abs(axis-1)], word.endPos[abs(axis-1)]+1))
            compPos = set(range(comp.startPos[abs(axis-1)], comp.endPos[abs(axis-1)]+1))

            if len(wordPos & compPos):
                if word.length > comp.length:
                    removeSet.add(comp)
                else:
                    removeSet.add(word)

# delete from list
for word in removeSet:
    foundWords.remove(word)

# output
print(len(foundWords))
for word in foundWords:
    print(word.startPos[0]+1, word.startPos[1]+1, word.endPos[0]+1, word.endPos[1]+1)

points = 0
for word in foundWords:
    points += word.length

print(points)
```
