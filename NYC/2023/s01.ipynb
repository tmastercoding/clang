```python
# 인류의 적 모기 퇴치
# NYPC 2023 · Round 1 연습문제
# 최근 이상 기후로 인해 붐힐 마을에 모기가 많이 늘어났다. 
# 마리드는 붐힐 마을 주민들을 돕기 위해 물풍선으로 모기들을 퇴치하려고 한다. 
# 마리드는 기존의 십자 모양의 물풍선만으로 모기를 효율적으로 퇴치할 수 없다는 것을 깨달았고, 
# 엑스 자 모양으로 터지는 물풍선을 새로 개발했다.

# 붐힐 마을은 N×N 크기의 
# 2차원 배열로 표현할 수 있고, 
# 배열의 각 칸에 존재하는 모기의 수가 주어진다.

# 마리드는 한 칸을 선택하여 물풍선을 놓고 터트린다. 
# 물풍선을 터트리면 범위 M으로 물줄기가 뻗어나가 모기를 퇴치할 수 있다.
# 예를 들어, 십자 모양의 경우, 
# 물풍선을 터트린 칸에서 위쪽, 오른쪽, 아래쪽, 왼쪽의 
# M 개 칸 이내의 모든 모기를 퇴치한다. 
# 엑스 자 모양의 경우, 네 개의 대각선 방향으로 
# M 개 칸 이내의 모든 모기를 퇴치한다.

# 마리드가 물풍선을 놓는 칸은 마을 안에 있어야만 하지만, 
# 물풍선이 터지는 물줄기가 마을 밖으로 나가는 경우는 허용된다.



# 위에서 왼쪽 그림은 
# M=3일 때 십자 모양의 물풍선이 터져 나오는 물줄기를 보여주고, 
# 오른쪽 그림은 
# M=3일 때 엑스 자 모양의 물풍선이 터져 나오는 물줄기를 보여준다.

# 십자 모양 혹은 엑스 자 모양의 물풍선을 한 개 터트릴 때, 
# 퇴치할 수 있는 모기 수의 최댓값을 계산하는 프로그램을 작성하시오.

# 입력 형식
# 첫 줄에 붐힐 마을의 크기를 나타내는 정수 
# N과 물풍선의 범위를 나타내는 정수 
# M이 공백으로 구분되어 주어진다. 
# (1≤N≤50; 0≤M≤50)

# 이어지는 N 개 줄의 
# i 번째 줄에는 붐힐 마을을 나타내는 배열의 
# i 번째 행의
# N 개의 칸에 존재하는 모기의 수를 나타내는 정수
# N 개가 공백으로 구분되어 순서대로 주어진다. 
# 이때, 주어지는 정수는 
# 0 이상 100 이하이다.

# 출력 형식
# 첫 줄에 십자 모양 혹은 엑스 자 모양의 물풍선을 한 개 터트릴 때, 
# 퇴치할 수 있는 모기 수의 최댓값을 출력한다.

# 예제 1
# 입력
# 3 1
# 1 2 1
# 2 2 2
# 1 2 1

# 출력
# 10

# 예제 2
# 입력
# 5 2
# 2 18 6 2 3
# 5 11 9 12 5
# 7 21 13 19 6
# 8 4 1 24 1
# 7 3 5 7 3

# 출력

# 104
```

```python

#
#     @ author :
#         - name  : Tay Oh
#         - email : tmastercubing@gmail.com
#     @ since : Sunday 9th June 2024
#     @ function name : get_dir
#     @ params : 
#         - arr: list                       : 2d array of mosquitoes
#         - point: tuple                    : coordinates of point
#         - size: int                       : length of row and col of town
#         - reach: int                      : how long the water splashes
#         - addToX: int                     : the val added to col (determines left or right)
#         - addToY: int                     : the val added to row (determines top or bottom)
#     @ return_value : sum of a certain direction
#     @ desc: |
#         gets a 2D arr of town and point, 
#         gets the sum of all mosquitoes
#         killed in one direction
#     @ function name : get_x
#     @ params : 
#         - arr: list                       : 2d array of mosquitoes
#         - point: tuple                    : coordinates of point
#         - size: int                       : length of row and col of town
#         - reach: int                      : how long the water splashes
#     @ return_value : sum of all directions in a X shape
#     @ desc: |
#         gets a 2D arr of town and point, 
#         gets the sum of all mosquitoes
#         killed in all directions
#     @ function name : cross
#     @ params : 
#         - point: tuple                    : coordinates of point
#         - arr: list                       : 2d array of mosquitoes
#         - until: int                      : how long the water splashes
#     @ return_value : sum of all directions in a + shape
#     @ desc: |
#         gets a 2D arr of town and point, 
#         gets the sum of all mosquitoes
#         killed in all directions
#

import numpy as np

size = 5
reach = 2

## get_left_top( point )
## row --, col --

def get_dir(arr: list, point: tuple, size, reach, addToX, addToY):
    row, col = point
    tot = 0
    count = 0
    while True:
        row += addToY
        col += addToX
        count+=1
        if not (0 <= row < size and 0 <= col < size):
            break

        if not(count < reach):
            break

        tot += arr[row][col]
    return tot

def get_x(arr: list, point: tuple, size, reach):
    return get_dir(arr, point, size, reach, 1, 1) \
    + get_dir(arr, point, size, reach, 1, -1)     \
    + get_dir(arr, point, size, reach, -1, -1)    \
    + get_dir(arr, point, size, reach, -1, 1)  \
    + arr[point[0]][point[1]]    

def cross( point, arr, until ):
    x, y = point
    arr = np.array( arr ) 
    tot = arr[x-until:x+until+1, y].sum() \
        + arr[x, y-until:y+until+1].sum() \
        - arr[x, y]
    
    return tot


test = [
[2, 18, 6, 2, 3],
[5, 11, 9, 12, 5],
[7, 21, 13, 19, 6],
[8, 4, 1, 24, 1],
[7, 3, 5, 7, 3]]

cross_list = []
for row in range( size ):
    for col in range( size ):
        cross_list.append( cross( (row, col), test, reach) )
        cross_list.append( get_x( test, (row, col), size, reach) )

print(max(cross_list))

# for idx in range(size):
#     for jdx in range(size):
#         totX = test[idx][jdx]
#         totX += get_right_top(test, (idx, jdx), size, reach)
#         totX += get_right_bottom(test, (idx, jdx),size,  reach)
#         totX += get_left_top(test, (idx, jdx), size, reach)
#         totX += get_left_bottom(test, (idx, jdx),size,  reach)

#         totPlus = test[idx][jdx]
#         totPlus += get_right(test, (idx, jdx), size, reach)
#         totPlus += get_bottom(test, (idx, jdx),size,  reach)
#         totPlus += get_top(test, (idx, jdx), size, reach)
#         totPlus += get_left(test, (idx, jdx), size, reach)

#         totList.append(totX)
#         totList.append(totPlus)

# print(totList)
```
