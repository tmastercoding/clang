```python
from itertools import combinations

# 선수 데이터 (포지션, 오버롤, 가격, 급여, 이름)
players = [
    ("GK", 20, 3, 5, "na"),
    ("GK", 30, 5, 10, "nb"),
    ("DF", 10, 2, 5, "nc"),
    ("DF", 10, 3, 5, "nc"),
    ("DF", 15, 3, 10, "nd"),
    ("DF", 15, 4, 10, "ne"),
    ("DF", 20, 5, 15, "nf"),
    ("DF", 20, 6, 15, "ng"),
    ("MF", 15, 2, 5, "nh"),
    ("MF", 15, 2, 5, "ni"),
    ("MF", 20, 3, 10, "nj"),
    ("MF", 20, 4, 15, "nk"),
    ("MF", 20, 5, 15, "nl"),
    ("MF", 20, 6, 20, "nl"),
    ("FW", 20, 10, 20, "nm"),
    ("FW", 30, 15, 30, "nm")
]

class Player:
    def __init__(self, pos, overall, price, salary, name):
        self.pos = pos
        self.overall = overall
        self.price = price
        self.salary = salary
        self.name = name

    def __str__(self):
        return (f'pos: {self.pos}\n'
                f'overall: {self.overall}\n'
                f'price: {self.price}\n'
                f'salary: {self.salary}\n'
                f'name: {self.name}')

# 선수 리스트 생성
playerList = [Player(*p) for p in players]

# 조건 정의
required_positions = {
    "GK": 1,
    "DF": (3, 5),
    "MF": (2, 5),
    "FW": (1, 4)
}

total_price_limit = 60
total_salary_limit = 120

def is_valid_combination(selected_players):
    position_count = {"GK": 0, "DF": 0, "MF": 0, "FW": 0}
    total_price = 0
    total_salary = 0
    total_overall = 0
    
    for player in selected_players:
        pos = player.pos
        if pos in position_count:
            position_count[pos] += 1
            total_price += player.price
            total_salary += player.salary
            total_overall += player.overall
            
    if total_price > total_price_limit or total_salary > total_salary_limit:
        return False
    
    if position_count["GK"] != required_positions["GK"]:
        return False
    
    if not (required_positions["DF"][0] <= position_count["DF"] <= required_positions["DF"][1]):
        return False
    
    if not (required_positions["MF"][0] <= position_count["MF"] <= required_positions["MF"][1]):
        return False
    
    if not (required_positions["FW"][0] <= position_count["FW"] <= required_positions["FW"][1]):
        return False
    
    return total_overall

best_overall_sum = 0

# 모든 선수 조합을 시도 (정확히 11명)
for comb in combinations(playerList, 11):
    overall_sum = is_valid_combination(comb)
    if overall_sum is not False and overall_sum > best_overall_sum:
        best_overall_sum = overall_sum

print(f"최대 오버롤의 합: {best_overall_sum}")
```
