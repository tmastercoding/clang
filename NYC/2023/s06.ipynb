```python
class Player():
    def __init__(self, pos, ovr, price, pay, name):
        self.pos = pos
        self.ovr = int(ovr)
        self.price = int(price)
        self.pay = int(pay)
        self.name = name

    def output(self):
        print(self.pos, self.ovr, self.price, self.pay, self.name)

case1 = '''16 60 
GK 20 3 5 na
GK 30 5 10 nb
DF 10 2 5 nc
DF 10 3 5 nc
DF 15 3 10 nd
DF 15 4 10 ne
DF 20 5 15 nf
DF 20 6 15 ng
MF 15 2 5 nh
MF 15 2 5 ni
MF 20 3 10 nj
MF 20 4 15 nk
MF 20 5 15 nl
MF 20 6 20 nl
FW 20 10 20 nm
FW 30 15 30 nm'''


limits = case1.split('\n')[0]
lines = case1.split('\n')[1:]

playerSize, budget = limits.split()
playerSize = int(playerSize)
budget = int(budget)

players = dict()

cnt = 1
for line in lines:
    players.update( { cnt: Player( *line.split() ) } )
    cnt +=1

# for key, val in players.items():
#     print(str(key) + ": ", end = '')
#     val.output()

max_size = 11

# GK = 1
# 3 <= DF <= 5
# 2 <= MF <= 5
# 1 <= FW <= 4

gk = list()
df = list()
mf = list()
fw = list()

for key, val in players.items():
    if val.pos == "GK":
        gk.append(val)
    if val.pos == "DF":
        df.append(val)
    if val.pos == "MF":
        mf.append(val)
    if val.pos == "FW":
        fw.append(val)
# print(gk)
# print(df)
# print(mf)
# print(fw)

priceCnt = 0
payCnt = 0
maxPay = 120
totOvr = 0
maxOvr = 0

# 90                                            
```
